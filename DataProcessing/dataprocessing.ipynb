```python
import pandas as pd
import numpy as np
# Load dataset
Daily_Data = pd.read_csv("daily-minimum-temperatures-in-me.csv")
print(Daily_Data.head())
```

```python
# ----------------------Data preprocessing-----------------------
## Part 1 Processing data style
# Tranverse the dates to datetime style
Daily_Data["Date"] = pd.to_datetime(Daily_Data["Date"], errors='coerce')
# Separate the year from dates
Daily_Data["Year"] = Daily_Data["Date"].dt.year
# Only keep month-day data
Daily_Data["Date"] = Daily_Data["Date"].dt.strftime("%m-%d")
## Part 2 Processing value style
Daily_Data["Value"] = Daily_Data["Value"].astype(float)
# Rearrange the order
Daily_Data = Daily_Data[["Year", "Date", "Value"]]
print(Daily_Data.head())
```

```python

uYear = Daily_Data['Year'].unique()
DictYear = {item:i for i, item in enumerate(uYear)}
uDate = Daily_Data['Date'].unique()
DictDate = {item:i for i, item in enumerate(uDate)}
print(DictYear)
print(DictDate)
```

```python
Index_1 = np.zeros((len(Daily_Data),1))
Index_2 = np.zeros((len(Daily_Data),1))
Results = []
for index, row in Daily_Data.iterrows():
    Index_1[index, 0] = DictYear[row["Year"]]
    Index_2[index, 0] = DictDate[row["Date"]]
    Results.append([Index_1[index, 0]+1, Index_2[index, 0]+1, row["Value"]])
```

```python
for i in Results:
    print("[%d, %d, %.1f]," % (i[0], i[1], i[2]))
```

```python
Daily_Data["x"] = Index_1
Daily_Data["y"] = Index_2
Daily_Data = Daily_Data[["Year", "Date", "x", "y", "Value"]]
print(Daily_Data.head())
Daily_Data.to_json("ProcessedData.json")
```
