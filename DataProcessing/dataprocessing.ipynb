```python
import pandas as pd
import numpy as np
# Load dataset
Daily_Data = pd.read_csv("daily-minimum-temperatures-in-me.csv")
print(Daily_Data.head())
```

```python
# ----------------------Data preprocessing-----------------------
## Part 1 Processing data style
# Tranverse the dates to datetime style
Daily_Data["Date"] = pd.to_datetime(Daily_Data["Date"], errors='coerce')
# Separate the year from dates
Daily_Data["Year"] = Daily_Data["Date"].dt.year
# Only keep month-day data
Daily_Data["Date"] = Daily_Data["Date"].dt.strftime("%m-%d")
## Part 2 Processing value style
Daily_Data["Value"] = Daily_Data["Value"].astype(float)
# Rearrange the order
Daily_Data = Daily_Data[["Year", "Date", "Value"]]
print(Daily_Data.head())
```

```python
Year = Daily_Data['Year'].unique()
DictYear = [item for item in Year]
Date = Daily_Data['Date'].unique()
DictDate = [item for item in Date]
print(DictYear)
print(DictDate)
```

```python
uYear = Daily_Data['Year'].unique()
DictYear = {item:i+1 for i, item in enumerate(uYear)}
uDate = Daily_Data['Date'].unique()
DictDate = {item:i+1 for i, item in enumerate(uDate)}
Results = []
for index, row in Daily_Data.iterrows():
    Results.append([DictYear[row["Year"]], DictDate[row["Date"]], row["Value"]])

for i in Results:
    print("[%d, %d, %.1f]," % (i[0], i[1], i[2]))
```

```python
## Experiment 3
# Load dataset
Data = pd.read_excel("CityData.xlsx")
# Tranverse the dates to datetime style
Data["updateTime"] = pd.to_datetime(Data["updateTime"])
Data["updateTime"] = Data["updateTime"].dt.strftime("%Y-%m-%d")
print(Data.head())
```

```python
# Produce daily data
def SelectData(Day):
   # Select the needed data 
   DailyData = Data.loc[Data["updateTime"] == Day, :]
   for i in Data["provinceName"].unique():
      ConfirmD = DailyData.loc[DailyData["provinceName"] == i, "city_confirmedCount"].sum()
      print("\t\t\t\t{'name': '%s', 'value': %d}," % (i, ConfirmD))
```

```python
# For task 1 we select the data until to "2020-04-01"
Task1_Day = "2020-04-01"
SelectData(Task1_Day)
```

```python
# For task 2
# Select the needed data
Date = []
for day in Data["updateTime"].unique():
    Date.append(day)
    print("{")
    print("\t\ttitle: '{}疫情数据',".format(day))
    print("\t\tseries: [{\n\t\t\tdata:\n\t\t\t[")
    SelectData(day)
    print("\t\t]\n\t}]\n},")


for i, day in enumerate(Date):
    print("'{}'".format(day), end=",")
    if (i+1) % 20 == 0:
        print()
```
